## Operation on edge detection

### importing lib


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### read image and view


```python
image = cv2.imread("/content/OIP (2).jpeg")
```

### Gray Scale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.axis('off')
plt.show()
```

```python
plt.imshow(edges)
plt.axis('off')
```

# Real time implementation 

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 30, 100)
    cv2.imshow("edges", edges)
    cv2.imshow("gray", gray)
    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

## Data gen for edge det

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("/content/drive/MyDrive/data/I joint") if isfile(join("/content/drive/MyDrive/data/I joint", f))]

print(onlyfiles)
```

### B&W analysis

```python
from google.colab.patches import cv2_imshow
destination = '/content/drive/MyDrive/BW/I-joint/'
for i in onlyfiles:
   path = '/content/drive/MyDrive/data/I joint/'+i
   img = cv2.imread(path)
   gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
   (thresh, binary) = cv2.threshold(gray_image, 150, 225, cv2.THRESH_BINARY)
   cv2_imshow(binary)
   cv2.imwrite(destination+i,img)

   
```

### Edge Detection 

```python
from google.colab.patches import cv2_imshow
destination = 'My_path_to_dir_with_file'
for i in onlyfiles:
   path = 'My_path_to_dir_with_fils'+i
   img = cv2.imread(path)
   gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
   edged_image = cv2.Canny(gray_image, threshold1=30, threshold2=100)
   cv2_imshow(edged_image)
   cv2.imwrite(destination+i,img)
```
